# Bank Marketing Dataset — Exploratory Data Analysis

### About the Dataset
The Bank Marketing Dataset is collected from a Portuguese 
banking institution. It contains data related to direct 
marketing campaigns conducted via phone calls. The goal 
of these campaigns was to convince customers to subscribe 
to a term deposit.

### Dataset Details
- Total Records: 11,162 customers
- Total Features: 17 columns
- Source: UCI Machine Learning Repository
- Target Column: deposit (yes/no)

### Column Description
- age: Age of the customer
- job: Type of job (management, technician, etc.)
- marital: Marital status (married, single, divorced)
- education: Education level (primary, secondary, tertiary)
- default: Has credit in default? (yes/no)
- balance: Average yearly balance in euros
- housing: Has housing loan? (yes/no)
- loan: Has personal loan? (yes/no)
- contact: Type of communication (cellular, telephone)
- day: Last contact day of the month
- month: Last contact month of the year
- duration: Last contact duration in seconds
- campaign: Number of contacts during this campaign
- pdays: Days since last contact from previous campaign
- previous: Number of contacts before this campaign
- poutcome: Outcome of previous campaign
- deposit: Has the client subscribed a term deposit? (yes/no)

### Objective
To explore and understand the dataset through statistical 
analysis and visualizations in order to find patterns that 
help predict which customers are likely to subscribe to a 
term deposit.

##### Checking for Outliers

```python
Q1 = df['balance'].quantile(0.25)
Q3 = df['balance'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['balance'] < lower) | (df['balance'] > upper)]
print(outliers)
print(f"Lower fence: {lower}")
print(f"Upper fence: {upper}")
print(f"Total outliers: {len(outliers)}")
```

```python
cols_to_check = ['balance', 'duration', 'campaign']

for col in cols_to_check:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outlier_count = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    print(f"{col}: {outlier_count} outliers detected — retained (realistic banking values)")
```

## Covariance Matrix

### Covariance vs Correlation

**Covariance** tells us if two columns move in the same direction or opposite 
directions — but the actual number is hard to judge on its own because it 
depends on the scale of the columns. For example, balance has much bigger 
numbers than age, so their covariance value comes out huge even if the actual 
relationship isn't that strong. This makes it tricky to compare covariance 
across different column pairs.

**Correlation** fixes this problem. It's basically the same idea as covariance, 
but standardized to always fall between -1 and +1, no matter what scale the 
original columns were in. That's why our correlation heatmap (made earlier) 
is much easier to read and compare — every value is on the same scale.

So in short: covariance is scale-dependent (affected by the size of numbers 
in each column), while correlation is scale-independent (always between -1 
and 1). That's also why we use correlation, not covariance, when comparing 
relationship strength across multiple column pairs.

The only moderately correlated pair, pdays-previous, is expected since both describe the customer's contact history.

## Overall Conclusion

### Summary of Key Findings

This EDA explored 11,162 bank marketing customers across 17 features, with the goal of understanding patterns related to term deposit subscription (the target variable, `deposit`).

**Target Variable**: The dataset is fairly balanced — 52.6% did not subscribe, 47.4% did — which is favorable for building a classification model without needing resampling techniques.

**Numerical Features**: All four key numeric columns (age, balance, duration, campaign) are right-skewed, but to very different degrees. Age is close to normal (skew = 0.86). Balance and campaign are extremely skewed (skew = 8.22 and 5.55) with sharp peaks and long tails, resembling log-normal and exponential distributions respectively — typical of financial and contact-count data. Outliers in these columns were retained rather than removed, since they represent realistic customer behavior (high balances, high contact attempts) rather than data errors.

**Categorical Features**: Most customers work in management, blue-collar, or technician roles; are married; and have secondary education. These represent the dominant customer profile in the dataset.

**Correlations Among Numeric Features**: Numeric features are largely independent of one another, with only pdays and previous showing a moderate relationship (0.51) — expected since both relate to prior campaign contact.

### Patterns Discovered

- **Duration is the strongest behavioral signal**: customers who subscribed had a noticeably higher median call duration (~430s) compared to those who didn't (~165s). This was the clearest pattern found in the entire analysis.
- **Balance shows a mild positive link to subscribing**, but it's not a strong standalone predictor.
- **Age shows almost no relationship** with either balance or subscription outcome — customers of all ages behave similarly.
- **Campaign and duration have a weak negative relationship** — customers contacted more times tend to have shorter calls, possibly indicating lower engagement on repeated attempts.

### Most Important Columns for Modeling

Based on this analysis, the features most likely to be useful for predicting `deposit` are:

1. **duration** — by far the strongest predictor identified in this EDA
2. **balance** — weak but present relationship with subscription
3. **campaign** — useful in combination with duration, may indicate diminishing returns from repeated contact
4. **job, education, marital** — useful categorical features for segmenting customer profiles, though not directly tested against `deposit` in this EDA

**age** and **pdays/previous** appear to have limited standalone predictive value based on the relationships explored here, though they may still contribute in combination with other features during modeling.
